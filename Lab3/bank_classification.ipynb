```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
```

# Data Loading

```python
# data loading

df = pd.read_csv('https://raw.githubusercontent.com/kovaciardey/ce888-repo/main/Lab3/bank-additional-full.csv', delimiter=";")

print("The dataset has %d columns and %d rows" % (df.shape[1], df.shape[0]))

df
```

# Data Prepocessing


```python
# dropping NA rows

print("Number of rows before dropping NaNs: %d" % len(df))
df = df.dropna()
print("Number of rows after dropping NaNs: %d" % len(df))
```

```python
# perform label encoder on the columns

headers = ["education", "month", "day_of_week", "poutcome", "y"]

df[headers] = df[headers].apply(LabelEncoder().fit_transform)

df.head()
```

```python
# create the dummies

df_dummies = pd.get_dummies(df)

df_dummies.head()
```

```python
# remove duration column

df.drop(['duration'], inplace=True, axis=1)

df.head()
```
